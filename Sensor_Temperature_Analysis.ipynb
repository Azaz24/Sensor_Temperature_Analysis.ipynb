```python
import numpy as np
data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])
```

# Task 1: Print the shape of data & Mean temperature

```python
print("shape of data:", data.shape)

mean_temp =  data.mean(axis=1)  # axis=1 is mean calculated row-wise (per station)

print("Mean temperature per staion:", mean_temp)
```

# Task 2: Extract Temperature Above 28°C



```python
mask = data > 28
high_temp = data [mask]
print("Temperature above 28°C:")
print(high_temp)
```

# Task 3: Normalize Data between 0 & 1

```python
# formula given: normalized = (data - data.min()) / (data.max() - data.min())

normalized = (data - data.min()) / (data.max() - data.min())

print("normalized dat:")   #Lowest temperature - 0
                           #Highest temperature - 1
print(normalized)
```
